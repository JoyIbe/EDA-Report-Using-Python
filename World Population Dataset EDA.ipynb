```python
#grouping by with aggregation for categorical data
df.groupby('Continent').mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)
```

```python
#grouping by with aggregation for categorical data, specifying unique numeric columns

df.groupby('Continent')[['2022 Population',
       '2020 Population', '2015 Population', '2010 Population',
       '2000 Population', '1990 Population', '1980 Population',
       '1970 Population']].mean().sort_values(by='2022 Population', ascending=False)
```

```python
#grouping by with aggregation for categorical data, short cut
# do: df3= df.groupby('Continent')[df.columns[5:13]].mean().sort_values(by='2022 Population', ascending=False)
df3 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean().sort_values(by='2022 Population', ascending=False)
df3
```

```python
df4 = df3.transpose()
df4
```

```python
plt.rcParams['figure.figsize'] = (20,7)
df4.plot()
```
